```python
from transformers import AutoTokenizer, AutoModel

# 1. Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("ZhangFuXi/Zh-MT-LLM", trust_remote_code=True, padding_side="left")
model = AutoModel.from_pretrained("ZhangFuXi/Zh-MT-LLM", trust_remote_code=True).half().cuda()

# 2. Set the model to evaluation mode
model = model.eval()

# 3. Start a chat
response, history = model.chat(tokenizer, "你好 (Hello)", history=[])
print(response)
```
